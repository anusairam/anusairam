**Exploring patterns and trends: A visual journey using matplotlib and seaborn.**

This is a data visualization skillfully crafted with matplotlib and seaborn. The visualization captivates with its blend of artistry and statistical insight, showcasing the beauty of data exploration through Python's powerful plotting libraries.Created different type of plots.

AREA CHARTS

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(0, 10, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

# Create Area Chart
plt.fill_between(x, y1, color='skyblue', alpha=0.4, label='Sin(x) Area')
plt.fill_between(x, y2, color='salmon', alpha=0.4, label='Cos(x) Area')

# Customize Plot (Optional)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Area Chart Example')
plt.legend()

# Display Plot
plt.show()
```

VIOLIN PLOT

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generates random data for the violin plot
np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=(100, 3))  # Three groups of data

# Creates a violin plot using Seaborn
plt.figure(figsize=(8, 6))
sns.violinplot(data=data, palette='Set3', inner='quartile')
plt.title('Violin Plot Example')
plt.xlabel('Groups')
plt.ylabel('Values')
plt.show()
```

FACET GRID

```python

import pandas as pd

# Create a DataFrame with sample data
np.random.seed(0)
data = pd.DataFrame({
    'X': np.random.randint(1, 10, 100),
    'Y': np.random.randint(1, 10, 100),
    'Category': np.random.choice(['A', 'B', 'C'], 100)
})

# Create a Facet Grid using Seaborn
g = sns.FacetGrid(data, col='Category', hue='Category', height=4)
g.map(plt.scatter, 'X', 'Y')
g.add_legend(title='Category')
plt.show()
```

PAIR PLOT

```python


# Creates a DataFrame with sample data
data = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [3, 1, 4, 2, 5],
    'Category': ['X', 'Y', 'Z', 'X', 'Y']
})

# Creates a Pair Plot using Seaborn
sns.pairplot(data, hue='Category', palette='Set1')
plt.title('Pair Plot Example')
plt.show()
```

```python

data = np.random.rand(10, 10)

# Creates a heatmap using Seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(data, cmap='viridis', annot=True, fmt=".2f", linewidths=.5)
plt.title('Heatmap Example')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

RADAR OR SPIDER CHART

```python


# Data
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4', 'Category 5']
values = [4, 3, 2, 5, 4]  # Values for each category
num_categories = len(categories)

# Create Polar Chart
angles = np.linspace(0, 2 * np.pi, num_categories, endpoint=False).tolist()
values += values[:1]  # Close the loop
angles += angles[:1]  # Close the loop

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, values, color='blue', alpha=0.25)  # Fill the area
ax.plot(angles, values, color='blue', linewidth=2)  # Plot the lines

# Customize Plot (Optional)
ax.set_theta_offset(np.pi / 2)  # Adjust angle offset
ax.set_theta_direction(-1)  # Change direction (clockwise or counterclockwise)
ax.set_ylim(0, 5)  # Set y-axis limit
ax.set_xticks(angles[:-1])  # Set x-axis ticks
ax.set_xticklabels(categories)  # Set x-axis labels

# Display Plot
plt.show()
```

ERROR BARS

```python
# Data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [20, 35, 30, 25]  # Corresponding values for each category
errors = [2, 3, 4, 1]  # Error values for each category (standard deviations, for example)

# Create Bar Plot with Error Bars
plt.bar(categories, values, yerr=errors, capsize=5)

# Customize Plot (Optional)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot with Error Bars Example')

# Display Plot
plt.show()
```

PIE CHARTS

```python


# Data
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [25, 30, 20, 25]  # Sizes or proportions for each category
colors = ['gold', 'lightblue', 'lightgreen', 'lightcoral']  # Colors for each category
explode = (0, 0.1, 0, 0)  # Explode one slice (Category B) for emphasis

# CreateS Pie Chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)


plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.title('Pie Chart Example')

# Display Plot
plt.show()
```

SUBPLOTS

```python


# Data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Create Subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot data on each subplot
axs[0, 0].plot(x, y1, 'b-', label='Sin(x)')
axs[0, 0].set_title('Subplot 1')
axs[0, 1].plot(x, y2, 'r-', label='Cos(x)')
axs[0, 1].set_title('Subplot 2')
axs[1, 0].plot(x, y3, 'g-', label='Tan(x)')
axs[1, 0].set_title('Subplot 3')
axs[1, 1].scatter(x, y1, color='orange', label='Sin(x) Scatter')
axs[1, 1].set_title('Subplot 4')

# Add legends to subplots
for ax in axs.flat:
    ax.legend()

# Adjust layout
plt.tight_layout()

# Display Plot
plt.show()


```

MULTIPLE AXES

```python


# Data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create Figure and Axes
fig, ax1 = plt.subplots()

# Plot data on first axes
ax1.plot(x, y1, 'b-')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y1-axis', color='b')

# Create a second y-axis on the right side
ax2 = ax1.twinx()
ax2.plot(x, y2, 'r-')
ax2.set_ylabel('Y2-axis', color='r')

# Customize Plot (Optional)
ax1.set_title('Multiple Axes Example')

# Display Plot
plt.show()
```
